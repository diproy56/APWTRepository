```python
class Boat:
    def __init__(self):
        # Starting values
        self.x = 0
        self.y = 0
        self.f = ""  # Direction the boat faces
        self.active = False
        self.limit = 4

    def depart(self, start_x, start_y, direction):
        if 0 <= start_x <= self.limit and 0 <= start_y <= self.limit:
            self.x = start_x
            self.y = start_y
            self.f = direction
            self.active = True

    def sail(self):
        nx, ny = self.x, self.y

        if self.f == "NORTH": ny -= 1
        elif self.f == "SOUTH": ny += 1
        elif self.f == "EAST": nx += 1
        elif self.f == "WEST": nx -= 1

        # Check if the next spot is safe before moving
        if 0 <= nx <= self.limit and 0 <= ny <= self.limit:
            self.x, self.y = nx, ny

    def rotate(self, turn_to):
        # A list of directions in order
        dirs = ["NORTH", "EAST", "SOUTH", "WEST"]
        idx = dirs.index(self.f)

        if turn_to == "STARBOARD":
            self.f = dirs[(idx + 1) % 4]
        elif turn_to == "PORT":
            self.f = dirs[(idx - 1) % 4]

    def status(self):
        # Show the result to the user
        print(f"{self.x},{self.y},{self.f}")

# Main Program
my_boat = Boat()

try:
    while True:
        text = input().strip()
        if not text:
            continue

        if text.startswith("DEPART"):
            try:
                # Grab the numbers and direction from the text
                info = text[7:]
                parts = info.split(",")
                my_boat.depart(int(parts[0]), int(parts[1]), parts[2].strip())
            except:
                pass

        elif my_boat.active:
            if text == "SAIL":
                my_boat.sail()
            elif text == "PORT":
                my_boat.rotate("PORT")
            elif text == "STARBOARD":
                my_boat.rotate("STARBOARD")
            elif text == "STATUS":
                my_boat.status()
except EOFError:
    pass
```
